```python
import os
from autogen import AssistantAgent
from autogen.agentchat.contrib.retrieve_user_proxy_agent import RetrieveUserProxyAgent
from autogen.retrieve_utils import extract_text_from_pdf

# 1. Configure the Ollama LLM Connection
# Assumes Ollama is running locally on port 11434 with llama3 or mistral pulled.
config_list = [
    {
        "model": "llama3.1",  # Replace with your specific Ollama model name
        "base_url": "http://localhost:11434/v1",
        "api_key": "ollama",  # Ollama does not require a real key
    }
]

llm_config = {
    "config_list": config_list,
    "cache_seed": 42,  # Cache results for faster testing
}

from pypdf import PdfReader
# Extract text from local PDF files
pdf_files = ["book.pdf"]
docs_content = PdfReader("book.pdf", strict=False)


#docs_content= extract_text_from_pdf("book.pdf")
```

```python
assistant = AssistantAgent(
    name="assistant",
    llm_config=llm_config,
    system_message="You are an expert assistant. If a calculation is required, write a complete Python code block."
)

ragproxyagent = RetrieveUserProxyAgent(
    name="ragproxyagent",
    human_input_mode="NEVER",
    max_consecutive_auto_reply=3,
    retrieve_config={
        "task": "qa",
        "docs_path": ["book.pdf"],
        "chunk_token_size": 200,
        "model": config_list[0]["model"],
        "vector_db": "chroma",
        "overwrite": True,
        "get_or_create": True,
    },
    code_execution_config=False,
)

assistant.reset()

qa_problem = "summarize the content and use python code to any calculation given in the book"

chat_result = ragproxyagent.initiate_chat(
    assistant,
    message=ragproxyagent.message_generator,
    problem=qa_problem
)
```
